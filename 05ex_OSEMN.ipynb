1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import numpy.random as npr
npr.seed(42)
with open("simple_data2.txt","+a") as f:
    a=npr.random((100,1))
    for i in a:
        f.write(str(i))

df=pd.DataFrame(a)
df.to_csv("simple_data.txt",index=False)
```
