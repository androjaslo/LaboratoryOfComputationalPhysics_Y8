3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello(func):
    def wrap(*args):
        print("Hello baby")
        func(*args)
        print("Funcion de funcion")
    return wrap

@hello
def square(x):
    return print(x**2)

print(square(3))
````
