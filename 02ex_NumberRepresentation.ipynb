1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
import numpy as np

def bin_dec(x):
    num=str(x)
    num=num[::-1]
    d=0
    for i in range(len(num)):
        d+=(int(num[i])*(2**(i)))
        #print(i,num[i],d)
    return d

    
print('To decimal: ',bin_dec(1010101))


def dec_bin(x):
    bi=[]
    while x!=1:
        bi.append(x%2)
        x=x//2
    if x==1:
        bi.append(1)
        
    if bi[0]==0:
        bi.insert(0,1)
    bi=bi[::-1]
    return np.asarray(bi)

print('To binary : ', dec_bin(85))



def hex_dec(x):
    num=str(x)
    num=num[::-1]  
    d=0
    for i in range(len(num)):
        if i=='A':
            i=10
        elif i=='B':
            i=11
        elif i=='C':
            i=12
        elif i=='D':
            i=13
        elif i=='E':
            i=14
        elif i=='F':
            i=15
        d+=(int(num[i])*(16**(i)))
        return d


print('Hex to dec : ',hex_dec(60))




def dec_hex(x):
    lista='0123456789ABCDEF'
    v=[]
    if x//16!=0:
        a=str(x%16)
        v.append(lista.index(a))
        c=x//16
        if c//16==0:
            a=str(c%16)
            v.append(lista.index(a))
        x=x//16
    v=v[::-1]
    return np.asarray(v)
print('Dec to hex : ', dec_hex(60))

```

7\. Consider the integral of the semicircle of radius 1:
$$
I=\int_{-1}^{1} \sqrt(1-x^2) {\rm d}x
$$
which it's known to be $I=\frac{\pi}{2}=1.57079632679...$.
Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compares to the true value?

(b) How much can $N$ be increased if the computation needs to be run in less than a second? What is the gain in running it for 1 minute? 


```python
import numpy as np

def f(x):
    return (1-x**2)**(0.5)
def riemann(n):
    x=np.linspace(-1,1,n)
    a=0
    h=2/n
    a=sum(f(x))*h

    return a
print(riemann(100))
#print(riemann(600000000))
%timeit riemann(100000)
%timeit riemann(1000000)
%timeit riemann(10000000)
%timeit riemann(100000000)
```
